```python
import matplotlib.pyplot as plt
```

```python
x=(1,2,3,4,5)
y=(3,6,9,12,15)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y)
plt.ylabel("idade")
plt.xlabel("carga")
```

```python
plt.plot(y,x)
plt.ylabel("idade")
plt.xlabel("carga")
```

```python
plt.plot(y,x)
plt.ylabel("idade")
plt.xlabel("carga")
plt.title("carga por idade")
```

```python
plt.plot(x,y, color="red")
plt.ylabel("idade")
plt.xlabel("carga")
plt.title("carga por idade")
```

```python
plt.bar(x,y, color="purple")
plt.ylabel("idade")
plt.xlabel("carga")
plt.title("carga por idade")
```

```python
plt.barh(x,y, color="purple")
plt.ylabel("idade")
plt.xlabel("carga")
plt.title("carga por idade")
```
